```python
import pandas as pd
import random

from numpy import array
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = pd.read_csv('Fortune500_financial_quarter_data_1961_2019.csv')
```

```python
some = data.drop('conm', 1).isna().groupby(data.conm, sort=False).sum().reset_index()
```

```python
other = some[some['prccq'] == 0]
```

```python
list(other['conm'].unique())
```

```python
companies = list(other['conm'].unique())
random25 = random.sample(companies, 25)
```

```python
random25
```

```python
hwdata = data.loc[data['conm'].isin(random25)] 
```

```python
hwdata = hwdata.reset_index(drop = True)
```

```python
hwdata
```

```python
hwdata.groupby('conm').count()
```

```python
hwdata.to_csv('hwdata.csv')
```
